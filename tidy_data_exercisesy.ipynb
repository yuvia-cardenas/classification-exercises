```python
import pandas as pd
import numpy as np

import matplotlib.pylab as plt
import seaborn as sns
import env

db_url = f'mysql+pymysql://{env.user}:{env.pwd}@{env.host}/tidy_data'
```

```python
sql_query = '''
                SELECT *
                From attendance;
                '''
attendance_df = pd.read_sql(sql_query, db_url )
```

```python
attendance_df
```

```python
# rename the problem column of Unnamed :0
attendance_df = attendance_df.rename(columns={'Unnamed: 0':'name'})
attendance_df
```

```python
attendance_df
```

```python
#melting the dates down resulting columns will be name date attend 
#     melt at id_vars = name the variable thats being the anchor
#     var_name is the name of the new column that contains the columns that collapse
#     value_name is columns name that holds current row values
attendance_df = attendance_df.melt(id_vars='name', var_name='date', value_name='attend')
attendance_df
```

```python
attendance_df.attend = attendance_df.attend.map({'P':1, 'H':.5, 'A':0, 'T':.9})
attendance_df
```

```python
attendance_df.groupby('name').mean()
```
